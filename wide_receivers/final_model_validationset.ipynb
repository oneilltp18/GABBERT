```python
## I want to create a list of the column names associated with a players
## first and second year stats, the following loop will go through and add
## all these stats to the list year_12

year_12 = []
for col in df:
    if col == 'year_1_growth':
        pass
    else:
        if '_0' in col:
            year_12.append(col)
        if '_1' in col:
            year_12.append(col)
            
            
## I also want to create a list of the rest of the col names (year3 data)
year_3 = []
for col in df:
    if col not in year_12:
        year_3.append(col)
        
year_3
```

```python
## I am going to repeat this process for the rookie data as well
rooksyear_12 = []
for col in rooks_df:
    if col == 'year_1_growth':
        pass
    else:
        if '_0' in col:
            rooksyear_12.append(col)
        if '_1' in col:
            rooksyear_12.append(col)
            
            
## I also want to create a list of the rest of the col names (year3 data)
rooksyear_3 = []
for col in rooks_df:
    if col not in rooksyear_12:
        rooksyear_3.append(col)
        
rooksyear_3
```

# PCA

```python
## I now want to perform PCA on all the year 1 and 2 stats columns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale

# I am creating a dataframe of the scaled values of the year 1 and 2 columns
pca_df = scale(df[year_12])

# doing the same thing for the rookie data
rookspca_df = scale(rooks_df[rooksyear_12])
```

```python
## creating the covariance matrix - this explains the variance between the different
## features within our dataframe

## for example, the value in the i,j position within the matrix explains the variance
## between the ith and the jth elements of a random vector, or between our features
cov_mat = np.cov(pca_df.T)

## creating my eigenvalues and corresponding eigenvectors
eigenValues, eigenVectors = np.linalg.eig(cov_mat)

## creating the eigenpairs - just pairing the eigenvalue with its eigenvector
eigenPairs = [(np.abs(eigenValues[i]), eigenVectors[:,i]) for i in range(len(eigenValues))]
```

```python
## loop through the eigenpairs and printing out the first row (eigenvalue)
## this is also seen in the code block above but just wanted to loop through again
## as it is a bit more clear like this
## I am also creating a list of the eigenvalues in ascending order to be able to reference it
sort_values = []
for i in eigenPairs:
    print i[0]
    sort_values.append(i[0])
```

```python
## we have the eigenvalues above showing us feature correlation explanation, but it helps
## to see the cumulative variance explained as well, which i can show below

## need to sum the eigen values to get percentages
sumEigenvalues = sum(eigenValues)

## this is a percentage explanation
variance_explained = [(i/sumEigenvalues)*100 for i in sort_values]
variance_explained
```

```python
### based on the above results, it seems that sticking to 25 features would be a decent
## cutoff point since the variance explained altogether reaches 85%

## this can very easily be manipulated by changing n_components adn then adding/subtracting
## columns to the dataframe in the code block below

## instantiate
pca = PCA(n_components = 25)

## fit and transform the standardized data
pca_model = pca.fit(pca_df)
pca_cols = pca_model.transform(pca_df)

## Here I am fitting the rookie data to the training PCA fit
rooks_pca_cols = pca_model.transform(rookspca_df)
```

```python
## Here I am simply creating the column headers for the pca features
pca_col_list = []

for i in range(1, 26):
    pca_col_list.append('pca_'+str(i))
    
## going to organize the columns into dataframe for organization
## I am going to set the index of our pca dataframe to the names 
## of the related player
pca_df = pd.DataFrame(pca_cols, index = df.index, columns = pca_col_list)
## doing the same for the rookie information
rookspca_df = pd.DataFrame(rooks_pca_cols, index = rooks_df.index, columns = pca_col_list)
##previewing dataframe
pca_df.head()
rookspca_df.head()
```

# Year 3 Stats Feature Selection

```python
from sklearn.feature_selection import chi2, SelectKBest
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import recall_score

## Defining my features and my target
X = df[year_3].drop(['categories', 'compilation_3'], axis = 1)
y = df['categories']

# Instantiate
kbest = SelectKBest(k=20)
kbest.fit(X,y)
# Show the feature importance for Kbest of 20
kbest_importance = pd.DataFrame(zip(X.columns, kbest.get_support()), columns = ['feature', 'important?'])

kbest_features = kbest_importance[kbest_importance['important?'] == True].feature

## creating a dataframe of just the best features from year 3 here
year_3_best = df[kbest_features]
rooks_year_3_best = rooks_df[kbest_features]
```

```python
## I now want to combine my PCA feature columns with my year 3 data to model
data = pca_df.join(year_3_best)
rooks_data = rookspca_df.join(rooks_year_3_best)
## defining my x and y again to perform select k best a second time
X = data
y = df['categories']


## I want to pick the 30 best features from the first select k best and pca cols
kbest_2 = SelectKBest(k = 30)
kbest_2.fit(X, y)

# Show the feature importance for Kbest of 30
kbest_2_importance = pd.DataFrame(zip(X.columns, kbest_2.get_support()), columns = ['feature', 'important?'])

kbest_2_features = kbest_2_importance[kbest_2_importance['important?'] == True].feature

## this dataframe is the best pca and year3 features together
year_3_refined = data[kbest_2_features]
rooks_year_3_refined = rooks_data[kbest_2_features]

print year_3_refined.shape
print rooks_year_3_refined.shape
rooks_year_3_refined.head()
```

# Model Construction

### SVC

```python
## I want to build an SVC model to grid search through based on my data df
from sklearn.svm import SVC
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.cross_validation import cross_val_predict


## Defining my features and my target
X = year_3_refined
y = df['categories']

# assign some category weights to try out
cat_weights = {'all_pro':3, 'below average' : .6, 'league_average' : 2,
                  'quality starter' : 5}

## Instantiating an empty SVC along with some others trying to tune params
svc = SVC()

c_svc = SVC(C = .7, class_weight = cat_weights,
              kernel = 'linear')

kernel_svc = SVC(kernel = 'poly', class_weight = 'balanced', degree = 3)
```

```python
cvp = cross_val_predict(c_svc, X, y, cv = 4, n_jobs = -1, verbose = 1)

print classification_report(y, cvp)
```

```python
cvp = cross_val_predict(kernel_svc, X, y, cv = 4, n_jobs = -1, verbose = 1)
print classification_report(y, cvp)
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

knn_params = dict(n_neighbors=range(1, 100), weights = ['distance', 'uniform'],
                 algorithm=['ball_tree', 'kd_tree', 'auto'])

knn_grid = GridSearchCV(knn, param_grid = knn_params, cv = 4,
                        scoring = 'recall_macro', n_jobs = -1, verbose = 1)
knn_fit = knn_grid.fit(X, y)

print knn_fit.best_score_
print
print knn_fit.best_estimator_
print
print knn_fit.best_params_

knn_ideal = knn_fit.best_estimator_

cvp_pred = cross_val_predict(knn_ideal, X, y, cv = 4, n_jobs = -1, verbose = 1)

print classification_report(y, cvp_pred)
```

### Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler

mnb = MultinomialNB()

X = MinMaxScaler(feature_range = (1, 6)).fit_transform(year_3_refined)
y = df['categories']

nb_params = dict(alpha=[.001, .005, .008, .01, .015, .02, .03, .04, .05, .1, .15, .2, .4, .6, .8, 1., 1.2, 1.4, 1.6])

nb_grid = GridSearchCV(mnb, param_grid = nb_params, cv = 4,
                      scoring = 'recall_macro', n_jobs = -1, verbose = 1)

nb_fit = nb_grid.fit(X, y)

print nb_fit.best_score_
print
print nb_fit.best_estimator_
print
print nb_fit.best_params_

nb_ideal = nb_fit.best_estimator_

nb_pred = cross_val_predict(nb_ideal, X, y, cv = 4, n_jobs = -1, verbose = 1)

print classification_report(y, nb_pred)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

cat_weights = {'all_pro':3, 'below average' : .6, 'league_average' : 2,
                  'quality starter' : 5}

lr_params = dict(penalty = ['l2'], C=[.1, .3, .5,.7,.9, 1., 1.1,1.3,1.5],
                class_weight = ['balanced'],
                solver = ['newton-cg', 'lbfgs'])
                 

lr_grid = GridSearchCV(lr, param_grid = lr_params, cv = 4, n_jobs = -1, verbose = 1)
lr_fit = lr_grid.fit(X, y)

print lr_fit.best_score_
print
print lr_fit.best_estimator_
print
print lr_fit.best_params_

lr_ideal = lr_fit.best_estimator_
lr_pred = cross_val_predict(lr_ideal, X, y, cv = 4, n_jobs = -1, verbose = 1)

print classification_report(y, lr_pred)
```

```python
allpro_weights = {'all_pro':4, 'below average' : .5, 'league_average' : .8,
                  'quality starter' : .8}

lr_ap = LogisticRegression(C = .5, class_weight = 'balanced', multi_class = 'ovr', penalty = 'l2',
                          solver = 'lbfgs')

allpro_pred = cross_val_predict(lr_ap, X, y, cv = 4, n_jobs = -1, verbose = 1)

print classification_report(y, allpro_pred)
```
